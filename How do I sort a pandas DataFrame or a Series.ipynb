# How do I sort a pandas DataFrame or a Series?

```python
import pandas as pd
```

```python
movies = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv")
```

```python
movies.head()
```

```python
movies.title.sort_values()
```

```python
type(movies['title'].sort_values())
```

```python
movies['title'].sort_values(ascending=False)
```

```python
movies.sort_values('star_rating', ascending=False)
```

```python
movies.sort_values(['actors_list', 'genre', 'star_rating'])
```
